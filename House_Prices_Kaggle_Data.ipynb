```python
rua_Street = {'Pave': 1, "Grvl": 0}
beco_Alley = {'Pave': 1, 'Grvl': 0}
utilidade_Utilities = {'AllPub': 1, 'NoSeWa': 0}
ar_CentralAir = {'Y': 1, 'N': 0}

df_train['Street'].replace(rua_Street, inplace = True)
df_train['Alley'].replace(beco_Alley, inplace = True)
df_train['Utilities'].replace(utilidade_Utilities, inplace = True)
df_train['CentralAir'].replace(ar_CentralAir, inplace = True)

df_train
```

```python
df_train_number = df_train.select_dtypes(exclude = 'object')
df_train_number.describe()
```

```python
(df_train_number.isnull().sum() / df_train_number.shape[0]).sort_values(ascending = False)
```

```python
df_train_number.drop("Alley", axis = 1, inplace = True)
df_train_number["LotFrontage"].fillna(0, inplace = True)
df_train_number.dropna(inplace=True)
```

```python
df_train_number["Bed/Rooms"] = df_train_number["BedroomAbvGr"] / df_train_number["TotRmsAbvGrd"]
df_train_number["BsmtFin1%"] = df_train_number["BsmtFinSF1"] / df_train_number["TotalBsmtSF"] 
df_train_number["BsmtFin2%"] = df_train_number["BsmtFinSF2"] / df_train_number["TotalBsmtSF"]
df_train_number["1stFlrArea%"] = df_train_number["1stFlrSF"] / df_train_number["LotArea"]
```

## Estimando Regressão (RFECV)

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_rfecv = RFECV(RandomForestClassifier(), scoring = 'accuracy',
                     step = 1)
```

```python
modelo_rfecv.score(X1, y1)
```
